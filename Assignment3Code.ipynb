```python
merge_MVP = MVP.drop(columns=['Year', 'League', 'WAR', 'SO'])
merge_MVP.head()
```

```python
newC = ['BA', 'OBP', 'SLG', 'HR', 'RBI', 'SB']
allP = pd.read_csv('mlb-player-stats-P.csv')
allPnew = allP.drop(columns=['Age', 'GS', 'G', 'CG', 'SHO', 'H', 'ER', 'BB', 'BS', 'HLD', 'WHIP'])
allPnew[newC] = None
allPnew.fillna(0.0, inplace=True)
allPnew.rename(columns={'Player': 'Name'}, inplace=True)
allPnew.head()
```
